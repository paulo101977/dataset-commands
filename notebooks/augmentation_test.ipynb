```python
import librosa
import pathlib
import os
import soundfile

import matplotlib.pyplot as plt
import numpy as np
import math
import ipynbname
from IPython import display
import wave

import torch
from audiomentations import *

nb_fname = ipynbname.name()
nb_path = ipynbname.path()

os.chdir('{0}/..'.format(os.path.split(nb_path)[0]))

if not os.path.exists("./augment"): 
    os.makedirs("./augment")

sample_file = './data/Direita_krikor.wav'

with wave.open(sample_file, "rb") as wave_file:
    frame_rate = wave_file.getframerate()

augment = AddBackgroundNoise(
    sounds_path="./noise/rain.wav",
    min_snr_in_db=0,
    max_snr_in_db=0.0,
    noise_transform=PolarityInversion(),
    p=1.0
)


signal, sr = librosa.load(sample_file, sr = frame_rate)


augmented_samples = augment(samples=signal, sample_rate=frame_rate)

soundfile.write('./augment/augmenttest.wav', augmented_samples, frame_rate)

display.display(display.Audio(signal, rate=frame_rate))
display.display(display.Audio(augmented_samples, rate=frame_rate))
```
